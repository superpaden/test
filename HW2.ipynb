```
%matplotlib inline
import numpy as up
import matplotlib.pyplot as plt
```

```
from tensorflow.keras.datasets import fashion_mnist
```

```
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D,MaxPooling2D
from tensorflow.keras.layers import Dense,Flatten
from tensorflow.keras.optimizers import SGD
```

```
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

```
%tensorflow_version 2.x
```

```
(x_train,y_train),(x_test,y_test) = fashion_mnist.load_data()
```

```
x_train = x_train.reshape(60000, 28, 28, 1) / 255
x_test = x_test.reshape(10000, 28, 28, 1) / 255
```

```
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)
```

```
x_test[1]
```

```
model = Sequential()
```

```
model.add(Conv2D(16,(4,4),padding="same",input_shape=(28,28,1),activation="elu" ))
```

```
model.add(MaxPooling2D(pool_size=(2,2)))
```

```
model.add(Conv2D(32, (3,3), padding='same',
                activation='selu'))
```

```
model.add(MaxPooling2D(pool_size=(2,2)))
```

```
model.add(Conv2D(64, (2,2), padding='same',
                activation='sigmoid'))
```

```
model.add(MaxPooling2D(pool_size=(2,2)))
```

```
model.add(Flatten())
```

```
model.add(Dense(99, activation='selu'))
```

```
model.add(Dense(10, activation='softmax'))
```

```
model.compile(loss='mse', optimizer=SGD(lr=0.087),
             metrics=['accuracy'])
model.fit(x_train, y_train, batch_size=128, epochs=12)
```

就是不用relu 改用relu的相似function(除了sigmoid) 不過sigmoid為甚麼不推薦 我查網上資料是說 不符合人神經的構造 而且 容易梯度消失 不知道為啥 開高走低


```
model1 = Sequential()
model1.add(Conv2D(16,(3,3),padding="same",input_shape=(28,28,1),activation="elu" ))
model1.add(MaxPooling2D(pool_size=(2,2)))
model1.add(Conv2D(32, (3,3), padding='same',
                activation='selu'))
model1.add(MaxPooling2D(pool_size=(2,2)))
model1.add(Conv2D(64, (3,3), padding='same',
                activation='elu'))
model1.add(MaxPooling2D(pool_size=(2,2)))
model1.add(Flatten())
model1.add(Dense(99, activation='selu'))
model1.add(Dense(10, activation='softmax'))
model1.compile(loss='mse', optimizer=SGD(lr=0.087),
             metrics=['accuracy'])
model1.fit(x_train, y_train, batch_size=128, epochs=12)
```

於是我修掉sigmoid 試試看  開超高走超低


```
model11 = Sequential()
model11.add(Conv2D(16,(3,3),padding="same",input_shape=(28,28,1),activation="relu" ))
model11.add(MaxPooling2D(pool_size=(2,2)))
model11.add(Conv2D(32, (3,3), padding='same',
                activation='relu'))
model11.add(MaxPooling2D(pool_size=(2,2)))
model11.add(Conv2D(64, (3,3), padding='same',
                activation='relu'))
model11.add(MaxPooling2D(pool_size=(2,2)))
model11.add(Flatten())
model11.add(Dense(99, activation='relu'))
model11.add(Dense(10, activation='softmax'))
model11.compile(loss='mse', optimizer=SGD(lr=0.087),
             metrics=['accuracy'])
model11.fit(x_train, y_train, batch_size=128, epochs=12)
```

放個全relu的對照組

```
model111 = Sequential()
model111.add(Conv2D(16,(3,3),padding="same",input_shape=(28,28,1),activation="elu" ))
model111.add(MaxPooling2D(pool_size=(2,2)))
model111.add(Conv2D(32, (3,3), padding='same',
                activation='selu'))
model111.add(MaxPooling2D(pool_size=(2,2)))
model111.add(Conv2D(64, (3,3), padding='same',
                activation='sigmoid'))
model111.add(MaxPooling2D(pool_size=(2,2)))
model111.add(Flatten())
model111.add(Dense(99, activation='selu'))
model111.add(Dense(10, activation='softmax'))
model111.compile(loss='mse', optimizer=SGD(lr=0.087),
             metrics=['accuracy'])
model111.fit(x_train, y_train, batch_size=128, epochs=12)
```

看來不用relu比較強 是因為relu比較容易死掉嗎

```
model1111 = Sequential()
model1111.add(Conv2D(16,(3,3),padding="same",input_shape=(28,28,1),activation="tanh" ))
model1111.add(MaxPooling2D(pool_size=(2,2)))
model1111.add(Conv2D(32, (3,3), padding='same',
                activation='sigmoid'))
model1111.add(MaxPooling2D(pool_size=(2,2)))
model1111.add(Conv2D(64, (3,3), padding='same',
                activation='sigmoid'))
model1111.add(MaxPooling2D(pool_size=(2,2)))
model1111.add(Flatten())
model1111.add(Dense(99, activation='tanh'))
model1111.add(Dense(10, activation='softmax'))
model1111.compile(loss='mse', optimizer=SGD(lr=0.087),
             metrics=['accuracy'])
model1111.fit(x_train, y_train, batch_size=128, epochs=12)
```
