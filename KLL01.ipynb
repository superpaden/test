```
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```
from tensorflow.keras.datasets import mnist
```

```
from tensorflow.keras.utils import to_categorical
```

```
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D,MaxPooling2D
from tensorflow.keras.layers import Dense,Flatten
from tensorflow.keras.optimizers import SGD,Adam
```

```
(x_train,y_train),(x_test,y_test) = mnist.load_data()
```

```
x_train = x_train.reshape(60000,28,28,1)/255
```

```
x_test = x_test.reshape(10000,28,28,1)/255
```

```
y_test = to_categorical(y_test,10)
y_train = to_categorical(y_train,10)
```

```
CNNrent= [Conv2D(16,(3,3),padding="same",input_shape=(28,28,1),activation="elu" ),MaxPooling2D(pool_size=(2,2)),Conv2D(32, (3,3), padding='same',
                activation='selu'),MaxPooling2D(pool_size=(2,2)),Conv2D(64, (3,3), padding='same',
                activation='elu'),MaxPooling2D(pool_size=(2,2))]
```

```
My = [Flatten(),Dense(99, activation='selu'),Dense(10, activation='softmax')]
```

```
for layer in CNNrent:
    layer.trainable = False
```

```
model = Sequential(CNNrent + My)
```

```
model.summary()
```

```
model.compile(loss="mse",optimizer=Adam(),metrics=["accuracy"])
```

```
model.load_weights("WTTT.h5")
```
