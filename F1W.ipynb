```
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```
from tensorflow.keras.datasets import fashion_mnist
```

```
(x_train,y_train),(x_test,y_test) = fashion_mnist.load_data()  #Q1 這個是預設好 60000 10000ㄇ? 沒有辦法改成其他比例嗎?
```

```
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

```
x_train = x_train.reshape(60000,784)
x_test=x_test.reshape(10000,784)
x_train = x_train/255
x_test = x_test/255
```

```
from tensorflow.keras.utils import to_categorical
```

```
y_copy = y_test
```

```
y_test = to_categorical(y_test,10)
y_train = to_categorical(y_train,10)
```

```
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD
```

```
model = Sequential() #為啥隨便改神經網路層 就有不同的結果 差異蠻大的
model.add(Dense(135,input_dim=784,activation="relu"))
model.add(Dense(84,activation="relu"))
model.add(Dense(39,activation="relu"))
model.add(Dense(14,activation="relu"))
model.add(Dense(10,activation="softmax"))
```

```
model.compile(loss="mse",optimizer=SGD(lr=0.0066),metrics=["accuracy"])
```

```
model.fit(x_train,y_train,batch_size=100,epochs=245) #預測245次就有80%
```

```
model.summary()
```

```
result = model.predict_classes(x_test)
```
